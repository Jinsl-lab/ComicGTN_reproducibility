```r
library(Matrix)
library(RaceID)
data <- readMM("./ComicGTN_reproducibility/Data/BMMC-bench-1/Gene_Cell.mtx")
Cell_names <- read.table("./ComicGTN_reproducibility/Data/BMMC-bench-1/Cell_names.tsv", header = FALSE)
Gene_names <- read.table("./ComicGTN_reproducibility/Data/BMMC-bench-1/Gene_names.tsv", header = FALSE)
rownames(data) <- Gene_names[,1]
colnames(data) <- Cell_names[,1]
```

```r
sc <- SCseq(data)
sc <- filterdata(sc,mintotal=300)
fdata <- getfdata(sc)
sc <- compdist(sc,metric="pearson")
sc <- clustexp(sc)
plotsaturation(sc,disp=FALSE)
plotjaccard(sc)
sc <- findoutliers(sc)
plotbackground(sc)
plotsensitivity(sc)
plotoutlierprobs(sc)
clustheatmap(sc)
sc <- comptsne(sc)
sc <- compfr(sc,knn=10)
sc <- compumap(sc)
plotmap(sc,um=TRUE)
```

```r
b <- prop.table(table(sc@cpart)) * 100
low_prop_categories <- b[b < 1]
result <- names(sc@cpart)
write.table(result, "./ComicGTN_reproducibility/Data/BMMC-bench-1/RaceID3_pred.tsv",sep = "\t",row.names = FALSE, col.names = FALSE)
```
