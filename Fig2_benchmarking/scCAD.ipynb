```python
import os
import scanpy as sc
import scCAD
import numpy as np
import pandas as pd
import h5py
from collections import Counter
```

```python
os.chdir("./ComicGTN_reproducibility/Data/BMMC-bench-1")

data_mat  = sc.read('Gene_Cell.mtx')
geneNames  = pd.read_csv('Gene_names.tsv', sep='\t', header=None)
cellNames  = pd.read_csv('Cell_names.tsv', sep='\t', header=None)
labels = pd.read_csv('Cell_types.tsv', sep='\t', header=None)

data = data_mat.X.transpose().toarray()
geneNames = np.array(geneNames[0])
cellNames = np.array(cellNames[0])
labels = np.array(labels[0])
```

```python
result, score, sub_clusters, degs_list = scCAD.scCAD(data=data, dataName='Sim', cellNames=cellNames, geneNames=geneNames, save_path='/home/jsl/YBR/scCAD/') 
```

```python
y_pred_binary = [1 if cellNames[i] in (result[0]) else 0 for i in range(len(cellNames))]
y_pred_binary = pd.DataFrame(y_pred_binary)
y_pred_binary.to_csv("./ComicGTN_reproducibility/Data/scCAD_pred.tsv", sep="\t", header=False, index=False)
```
