```r
library(DOSE)
library(enrichplot)
library(clusterProfiler)


path <- "./ComicGTN_reproducibility/Data/MOGHE_brain_cortex/Enrichment_results"
file_path1 <- paste(path, "Cluster_1_differential_genes.csv", sep = "/")
cluster1_DEG_all = read.table(file_path1, header = TRUE, sep = ",")
cluster1_DEG_ori = data.frame(symbol = cluster1_DEG_all$names[1:10])
cluster1_DEG <- bitr(cluster1_DEG_ori$symbol, fromType = "SYMBOL", toType = c("ENTREZID", "ENSEMBL"), 
                                  OrgDb = "org.Hs.eg.db", drop = FALSE)
cluster1_DEG = cluster1_DEG[is.na(cluster1_DEG[, "ENTREZID"]) == F, ]
result1 <- enrichDO(gene = cluster1_DEG$ENTREZID, ont = "DO", pvalueCutoff = 0.05, qvalueCutoff = 0.05, minGSSize = 1, readable = TRUE)
write.table(result1, file = "./ComicGTN_reproducibility/Data/MOGHE_brain_cortex/DO_results/cluster1_DO.txt", sep = "\t", quote = F, row.names = F)


file_path2 <- paste(path, "Cluster_2_differential_genes.csv", sep = "/")
cluster2_DEG_all = read.table(file_path2, header = TRUE, sep = ",")
cluster2_DEG_ori = data.frame(symbol = cluster2_DEG_all$names[1:10])
cluster2_DEG <- bitr(cluster2_DEG_ori$symbol, fromType = "SYMBOL", toType = c("ENTREZID", "ENSEMBL"), 
                                  OrgDb = "org.Hs.eg.db", drop = FALSE)
cluster2_DEG = cluster2_DEG[is.na(cluster2_DEG[, "ENTREZID"]) == F, ]
result2 <- enrichDO(gene = cluster2_DEG$ENTREZID, ont = "DO", pvalueCutoff = 0.05, qvalueCutoff = 0.05, minGSSize = 1, readable = TRUE)
write.table(result2, file = "./ComicGTN_reproducibility/Data/MOGHE_brain_cortex/DO_results/cluster2_DO.txt", sep = "\t", quote = F, row.names = F)


file_path3 <- paste(path, "Cluster_4_differential_genes.csv", sep = "/")
cluster4_DEG_all = read.table(file_path3, header = TRUE, sep = ",")
cluster4_DEG_ori = data.frame(symbol = cluster4_DEG_all$names[1:10])
cluster4_DEG <- bitr(cluster4_DEG_ori$symbol, fromType = "SYMBOL", toType = c("ENTREZID", "ENSEMBL"), 
                                  OrgDb = "org.Hs.eg.db", drop = FALSE)
cluster4_DEG = cluster4_DEG[is.na(cluster4_DEG[, "ENTREZID"]) == F, ]
result3 <- enrichDO(gene = cluster4_DEG$ENTREZID, ont = "DO", pvalueCutoff = 0.05, qvalueCutoff = 0.05, minGSSize = 1, readable = TRUE)
write.table(result3, file = "./ComicGTN_reproducibility/Data/MOGHE_brain_cortex/DO_results/cluster4_DO.txt", sep = "\t", quote = F, row.names = F)


file_path4 <- paste(path, "Cluster_7_differential_genes.csv", sep = "/")
cluster7_DEG_all = read.table(file_path4, header = TRUE, sep = ",")
cluster7_DEG_ori = data.frame(symbol = cluster7_DEG_all$names[1:10])
cluster7_DEG <- bitr(cluster7_DEG_ori$symbol, fromType = "SYMBOL", toType = c("ENTREZID", "ENSEMBL"), 
                                  OrgDb = "org.Hs.eg.db", drop = FALSE)
cluster7_DEG = cluster7_DEG[is.na(cluster7_DEG[, "ENTREZID"]) == F, ]
result4 <- enrichDO(gene = cluster7_DEG$ENTREZID, ont = "DO", pvalueCutoff = 0.05, qvalueCutoff = 0.05, minGSSize = 1, readable = TRUE)
write.table(result4, file = "./ComicGTN_reproducibility/Data/MOGHE_brain_cortex/DO_results/cluster7_DO.txt", sep = "\t", quote = F, row.names = F)


file_path5 <- paste(path, "Cluster_13_differential_genes.csv", sep = "/")
cluster13_DEG_all = read.table(file_path5, header = TRUE, sep = ",")
cluster13_DEG_ori = data.frame(symbol = cluster13_DEG_all$names[1:10])
cluster13_DEG <- bitr(cluster13_DEG_ori$symbol, fromType = "SYMBOL", toType = c("ENTREZID", "ENSEMBL"), 
                                    OrgDb = "org.Hs.eg.db", drop = FALSE)
cluster13_DEG = cluster13_DEG[is.na(cluster13_DEG[, "ENTREZID"]) == F, ]
result5 <- enrichDO(gene = cluster13_DEG$ENTREZID, ont = "DO", pvalueCutoff = 0.05, qvalueCutoff = 0.05, minGSSize = 1, readable = TRUE)
write.table(result5, file = "./ComicGTN_reproducibility/Data/MOGHE_brain_cortex/DO_results/cluster13_DO.txt", sep = "\t", quote = F, row.names = F)
```
